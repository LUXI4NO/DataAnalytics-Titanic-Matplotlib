```python
# Filtra el DataFrame para obtener las edades de los pasajeros que sobrevivieron (Survived == 1)
survived_ages = df[df['Survived'] == 1]['Age']

# Filtra el DataFrame para obtener las edades de los pasajeros que no sobrevivieron (Survived == 0)
deceased_ages = df[df['Survived'] == 0]['Age']
```

```python
# Creamos una figura con un tamaÃ±o especÃ­fico y una resoluciÃ³n determinada
fig = plt.figure(figsize=(1200 / 100, 675 / 100), dpi=140, facecolor="#EFE9E6")
ax = plt.subplot(111, facecolor="#EFE9E6")

# Ocultamos los bordes superior y derecho del grÃ¡fico
ax.spines["top"].set(visible=False)
ax.spines["right"].set(visible=False)

# AÃ±adimos una rejilla al grÃ¡fico con un estilo especÃ­fico y un color
ax.grid(True, color="lightgrey", ls=":")

# Definimos los intervalos de los bins para el histograma
bins = range(0, int(df['Age'].max()) + 20, 20)

# Creamos el histograma con los datos de las edades de los sobrevivientes y fallecidos
ax.hist([survived_ages, deceased_ages], bins=bins, color=['#FFDB58', '#E2725B'], ec='white', lw=0.75, label=['Sobrevivientes', 'Fallecidos'], zorder=3)

# Agregamos las curvas de densidad de distribuciÃ³n (KDE)
sns.kdeplot(survived_ages, color='#FFDB58', linestyle='-', label='KDE Sobrevivientes')
sns.kdeplot(deceased_ages, color='#E2725B', linestyle='-', label='KDE Fallecidos')

# AÃ±adimos una leyenda
ax.legend(['Sobrevivientes', 'Fallecidos'], ncol=2, loc="upper center", bbox_to_anchor=(0.46, 1.1), frameon=False)

# Etiquetamos las alturas de las barras en el histograma
for rect in ax.patches:
    height = rect.get_height()
    ax.annotate(f'{height:.0f}', xy=(rect.get_x() + rect.get_width() / 2, height), 
                xytext=(0, 3), textcoords="offset points", ha='center', va='bottom', 
                size=8, color='black', fontweight='bold')

# Configuramos las etiquetas del eje x e y
ax.set_xlabel('Edad', fontsize=12, fontname='Calibri', fontweight='bold', color='black')
ax.set_ylabel('Supervivientes', fontsize=12, fontname='Calibri', fontweight='bold', color='black')

# AÃ±adimos un tÃ­tulo a la figura
fig_text(
    x=0.5, y=0.990,
    s="DistribuciÃ³n de Edades de Sobrevivientes y Fallecidos en el Titanic",
    ha="center",
    family="Calibri",
    weight="bold",
    size=20
)

# Formateamos las etiquetas del eje x para que muestren solo nÃºmeros enteros
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{x:.0f} AÃ±os"))

# Configuramos las etiquetas del eje y
for label in ax.get_yticklabels():
    label.set_fontsize(8)
    label.set_color('black')
    label.set_fontname('sans-serif')
    label.set_fontweight('black')

# Configuramos las etiquetas del eje x
for label in ax.get_xticklabels():
    label.set_fontsize(8)
    label.set_color('black')
    label.set_fontname('sans-serif')
    label.set_fontweight('black')

# Calculamos y dibujamos las lÃ­neas horizontales para las edades promedio de los sobrevivientes y fallecidos
mean_survived = np.mean(survived_ages)
mean_deceased = np.mean(deceased_ages)
plt.axhline(mean_survived, color='#FFDB58', linestyle='--', label='Promedio Sobrevivientes')
plt.axhline(mean_deceased, color='#E2725B', linestyle='--', label='Promedio Fallecidos')

# Mostramos la figura
plt.show()
```
